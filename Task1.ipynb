```python
import pandas as pd
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
columns = ["sepal length",'sepal width',"petal length","petal width", "class"]
df = pd.read_csv("iris.csv", names = columns)
df
```

```python
df.describe()
```

```python
X = df.drop("class",axis=1)
Y = df["class"]
```

```python
Y
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
svm = LinearSVC()
svm.fit(X_train, Y_train)
```

```python
y_pred = svm.predict(X_test)
y_pred
```

```python
rforest = RandomForestClassifier()
rforest.fit(X_train,Y_train)
```

```python
rforest_pred = rforest.predict(X_test)
```

```python
print("The Classification report of Linear SVM for the IRIS DATASET IS SHOWN BELOW: \n",classification_report(Y_test,y_pred))
print("The Classification report of Random Forest for the IRIS DATASET IS SHOWN BELOW: \n",classification_report(Y_test,rforest_pred))
```

```python
from catboost import CatBoostClassifier
catmodel = CatBoostClassifier()
catmodel.fit(X_train,Y_train)
catmodel_pred = catmodel.predict(X_test)
```

```python
print("The report of Linear SVM for the IRIS DATASET IS SHOWN BELOW: \n",classification_report(Y_test,catmodel_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)
logreg_pred = logreg.predict(X_test)
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train,Y_train)
naive_bayes_pred = naive_bayes.predict(X_test)
```

```python
print("The classification report of Logistic Regression  for the IRIS DATASET IS SHOWN BELOW: \n",classification_report(Y_test,logreg_pred))
print("The classification report of Naive Bayes for the IRIS DATASET IS SHOWN BELOW: \n",classification_report(Y_test,naive_bayes_pred))
```
