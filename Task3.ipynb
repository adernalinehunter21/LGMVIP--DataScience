```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
df = df.sample(30000)
df
```

```python
df.info()
```

```python
songs_df = pd.read_csv("songs.csv")
songs_df
```

```python
df = pd.merge(df, songs_df, on='song_id', how='left')
del songs_df
```

```python
df.info()
```

```python
members_df = pd.read_csv("members.csv")
members_df
```

```python
df = pd.merge(df, members_df, on='msno', how='left')
del members_df
```

```python
df.info()
```

```python
for i in df.select_dtypes(include=['object']).columns:
    df[i][df[i].isnull()] = 'unknown'
df = df.fillna(value=0)
```

```python
df.info()
```

```python
df.registration_init_time = pd.to_datetime(df.registration_init_time, format='%Y%m%d', errors='ignore')
df['registration_init_time_year'] = df['registration_init_time'].dt.year
df['registration_init_time_month'] = df['registration_init_time'].dt.month
df['registration_init_time_day'] = df['registration_init_time'].dt.day

# expiration_date
df.expiration_date = pd.to_datetime(df.expiration_date,  format='%Y%m%d', errors='ignore')
df['expiration_date_year'] = df['expiration_date'].dt.year
df['expiration_date_month'] = df['expiration_date'].dt.month
df['expiration_date_day'] = df['expiration_date'].dt.day
```

```python
df['registration_init_time'] = df['registration_init_time'].astype('category')
df['expiration_date'] = df['expiration_date'].astype('category')
```

```python
# Object data to category
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')
    
# Encoding categorical features
for col in df.select_dtypes(include=['category']).columns:
    df[col] = df[col].cat.codes
```

```python
df.info()
```

```python
df.corr()
```

```python
plt.figure(figsize=[7,5])
sns.heatmap(df.corr())
plt.show()
```

```python
from sklearn.svm import LinearSVC
from  sklearn.linear_model import LogisticRegression
svm_model = LinearSVC()
svm_model.fit(df[df.columns[df.columns != 'target']], df.target)
```

```python
logreg = LogisticRegression()
logreg.fit(df[df.columns[df.columns != 'target']], df.target)
```

```python
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier()
adaboost.fit(df[df.columns[df.columns != 'target']], df.target)
```

```python
df_plot = pd.DataFrame({'features': df.columns[df.columns != 'target'],
                        'importances': adaboost.feature_importances_})
df_plot = df_plot.sort_values('importances', ascending=False)
```

```python
plt.figure(figsize=[11,5])
sns.barplot(x = df_plot.importances, y = df_plot.features)
plt.title('Importances of Features Plot')
plt.show()
```

```python
adaboost.feature_importances_
```

```python
df = df.drop(df_plot.features[df_plot.importances < 0.04].tolist(), 1)
```

```python
list(df.columns)
```

```python
target = df.pop('target')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(df, target, test_size = 0.3)
logreg.fit(train_data,train_labels)
```

```python
svm_model.fit(train_data,train_labels)
```

```python
from sklearn import metrics
```

```python
predict_labels = svm_model.predict(test_data)
predict_labels_logreg = logreg.predict(test_data)
print(metrics.classification_report(test_labels, predict_labels))
print(metrics.classification_report(test_labels, predict_labels_logreg))
```

```python
model = AdaBoostClassifier(learning_rate=0.1, n_estimators=250)
model.fit(train_data,train_labels)
```

```python
predict_adaboost_labels = model.predict(test_data)
print(metrics.classification_report(test_labels, predict_adaboost_labels))
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
xgbmodel = XGBClassifier()
gradboostmodel = GradientBoostingClassifier()
```

```python
xgbmodel.fit(train_data,train_labels)
```

```python
gradboostmodel.fit(train_data,train_labels)
```

```python
predict_gradboostmodel_labels = gradboostmodel.predict(test_data)
print(metrics.classification_report(test_labels, predict_gradboostmodel_labels))
```

```python
predict_xgbmodel_labels = xgbmodel.predict(test_data)
print(metrics.classification_report(test_labels, predict_xgbmodel_labels))
```
